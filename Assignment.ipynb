## Data cleaning

```python
#Drop SRAD column
data.drop('SRAD', axis=1, inplace=True)
```

```python
data_avg = data.groupby('DATE').mean()
data_avg
```

```python
data_avg.isna().sum()
```

```python
print(data_avg.TAVE.mean())
print(data_avg.TMIN.mean())
print(data_avg.TMAX.mean())
print(data_avg.RAIN.mean())
```

```python
data_avg.describe()
```

### Replace NaN value with the mean value of each column except 'RAIN' because it has more SD than other columns.

```python
data_avg.TAVE = data_avg.TAVE.fillna((data_avg.TAVE.mean()))
data_avg.TMIN = data_avg.TMIN.fillna((data_avg.TMIN.mean()))
data_avg.TMAX = data_avg.TMAX.fillna((data_avg.TMAX.mean()))
```

```python
data_avg.isna().sum()
```

```python
data_avg.shape
```

### remove N/A value 

```python
data_avg = data_avg.dropna()
data_avg.shape
```
