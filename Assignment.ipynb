## d - Plot the Fourier transform of TAVE. Do you notice some pattern ? How can you interpret it ?

```python
%matplotlib inline
from scipy.fftpack import fft
import scipy as sp
import scipy.fftpack

# Number of sample points
temp = data_avg['TAVE']
N = len(temp)
T = 1.0 / 365 #because the original unit is in days

yf = fft(temp.values)
xf = 1/(N*T)*np.arange(N//2)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
temp.plot(ax=ax, lw=.5)
ax.grid()
ax.set_xlabel('Date')
ax.set_ylabel('TAVE')
```

```python
# Compute the Fourier transform and the spectral density of the signal
# The first step is to compute the FFT of the signal using the fft() function:
temp_fft = sp.fftpack.fft(temp)

#Once the FFT has been obtained, we need to take the square of its absolute value in order to get the power spectral density (PSD):
temp_psd = np.abs(temp_fft) ** 2

#The next step is to get the frequencies corresponding to the values of the PSD using the fftfreq() utility function.Here, we choose an annual unit: a frequency of 1 corresponds to 1 year (365 days). We provide 1/365 because the original unit is in days:
fftfreq = sp.fftpack.fftfreq(len(temp_psd), 1. / 365)

#The fftfreq() function returns positive and negative frequencies. We are only interested in positive frequencies here, as we have a real signal
i = fftfreq > 0
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
ax.plot(fftfreq[i], 10 * np.log10(temp_psd[i]))
ax.grid()

ax.set_xlim(0, 5)
ax.set_ylim(15, 80)
ax.set_xlabel('Frequency (1/year)')
ax.set_ylabel('dB / (1/year)')
```

### Answer
#### The temperature in this real world, is still mostly dominated by “random” temperature levels.
<ul>
    <li>There is a peak in the graph at approximately 1 (cycle/year), which indicates a period of approximately 1 year</li>
</ul>

## e - Calculate the average probability of rain  (RAIN > 0) in 2000.

```python
#Find the total number of days in 2000
from datetime import date
import scipy
year = 2000
d0 = date(year, 1, 1)
d1 = date(year, 12, 31)
delta = (d1 - d0)
total_days = int(delta.days) + 1

group_YD = []
for day in range(total_days):  
    group_YD.append(str(year) + str("{0:0=3d}".format(day+1)))
```

```python
rainIn2000 = data_avg[['RAIN']].copy()

rainIn2000.drop(rainIn2000[(rainIn2000.RAIN <= 0) | ~rainIn2000.index.isin(group_YD)].index, inplace=True)
# rainIn2000
```

```python
mu=rainIn2000.RAIN.mean()
sig=rainIn2000.RAIN.std()
# rainIn2000['z']=scipy.stats.mstats.zscore(rainIn2000.RAIN)
rainIn2000['prob']=rainIn2000.RAIN.apply(lambda x: scipy.stats.norm(mu,sig).pdf(x) if x > mu else 1 - scipy.stats.norm(mu,sig).pdf(x))
```

### Answer

```python
print("The average probability of rain (RAIN > 0) in 2000 is {}".format(rainIn2000.prob.mean()))
```

## f - Calculate the median TMIN for October (calculated based on observations of any year as long as it is in October).

```python
def findOctInYear(year):
    delta_f = (date(year, 12, 31) - date(year, 1, 1))
    total_days_f = int(delta_f.days) + 1

    d_oct = date(year, 10, 1)
    d_dec = date(year, 12, 31)

    total_oct = date(year, 10, 31) - date(year, 10, 1)
    total_oct = int(total_oct.days) + 1

    delta_oct = (d_dec - d_oct)
    start_days_oct = (total_days_f - int(delta_oct.days))
    group_Oct = []
    
    for day in range(start_days_oct, start_days_oct + total_oct):  
        group_Oct.append(str(year) + str("{0:0=3d}".format(day)))    
    return group_Oct
```

```python
df_tmin = pd.DataFrame(columns =['MEDIAN'])
```

```python
for year in range(1988, 2019):
    oct_list = findOctInYear(year)
    median = data_avg.TMIN[data_avg.index.isin(oct_list)].median()
    df_tmin.loc[str(year)] = median
```
